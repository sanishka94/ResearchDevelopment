## 4.1 Social Network Data

```python
def read_graph(filename):
    G=nx.Graph()
    array = np.loadtxt(filename, dtype=int)
    G.add_edges_from(array)
    return G

fb = read_graph('facebook_combined.txt.gz')
n = len(fb)
m = len(fb.edges)
n, m
```

```python
average_clustering(fb, trials=1000)
```

```python
def sample_path_lengths(G, nodes=None, trials=1000):
    if nodes is None:
        nodes = list(G)
    else:
        nodes = list(nodes)
    
    pairs = np.random.choice(nodes, (trials, 2))
    lengths = [nx.shortest_path_length(G, *pair) for pair in pairs]
    return lengths

def estimate_path_lengths(G, nodes=None, trials=1000):
    return np.mean(sample_path_lengths(G, nodes, trials))

C = average_clustering(fb)
L = estimate_path_lengths(fb)
print('C', C)
print('L', L)
```

## Generating BA Graphs 

```python
def random_subset(repeated_nodes, k):
    targets = set()
    while len(targets) < k:
        x = random.choice(repeated_nodes)
        targets.add(x)
    return targets


def barabasi_albert_graph(n, k):
    G = nx.empty_graph(k)
    targets = list(range(k))
    repeated_nodes = []
    
    for source in range(k, n):
        G.add_edges_from(zip([source]*k, targets))
        
        repeated_nodes.extend(targets)
        repeated_nodes.extend([source]*k)
        
        targets = random_subset(repeated_nodes, k)
    
    return G
```

## Cumulative Distributions

```python
def cumulative_prob(pmf, x):
    ps = [pmf[value] for value in pmf if value<=x]
    return np.sum(ps)
```

```python
import gzip
```

```python
def all_pairs(nodes):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
            if i < j:
                yield u, v

def read_actor_network(filename, n=None):
    G = nx.Graph()
    with gzip.open(filename) as f:
        for i, line in enumerate(f):
            nodes = [int(x) for x in line.split()]
            G.add_edges_from(all_pairs(nodes))
            if n and i >= n:
                break
                
    return G
```

```python
actors = read_actor_network('actor.dat.gz')
```

```python
len(actors)
```

```python
deg_act = degrees(actors)
np.mean(deg_act), np.std(deg_act), average_clustering(actors)
```

```python
pmf_act = Pmf(deg_act)
cdf_act = Cdf(deg_act)

thinkplot.preplot(cols=2)
thinkplot.Pdf(pmf_act)

thinkplot.subplot(2)
thinkplot.Cdf(cdf_act)
```
