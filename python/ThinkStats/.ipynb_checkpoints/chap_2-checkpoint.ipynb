```python
preg = nsfg.ReadFemPreg()
live = preg[preg.outcome==1]
```

```python
hist = thinkstats2.Hist(live.birthwgt_lb, label='birthwgt_lb')
thinkplot.Hist(hist)
thinkplot.show(xlabel='pounds', ylabel='frequency')
```

```python
hist2 = thinkstats2.Hist(live.birthwgt_oz, label='birthwgt_oz')
thinkplot.Hist(hist2)
thinkplot.show(xlabel='ounces', ylabel='frequency')
```

```python
hist3 = thinkstats2.Hist(live.agepreg, label='agepreg')
thinkplot.Hist(hist3)
thinkplot.show(xlabel='age in years', ylabel='frequency')

hist4 = thinkstats2.Hist(live.prglngth, label='prglngth')
thinkplot.Hist(hist4)
thinkplot.show(xlabel='weeks', ylabel='frequency')
```

```python
for weeks, freq in hist4.Smallest(10):
    print(weeks, freq)
```

```python
for weeks, freq in hist4.Largest(7):
    print(weeks, freq)
```

```python
firsts = live[live.birthord==1]
others = live[live.birthord!=1]
firsts_hist = thinkstats2.Hist(firsts.prglngth)
others_hist = thinkstats2.Hist(others.prglngth)
```

```python
width = 0.45
thinkplot.PrePlot(2)
thinkplot.Hist(firsts_hist, align='right', width=width)
thinkplot.Hist(others_hist, align='left', width=width)
thinkplot.show(xlabel='weeks', ylabel='frequency', xlim=[27, 46])
```

```python
mean = live.prglngth.mean()
var = live.prglngth.var()
std = live.prglngth.std()
mean, var, std
var == std*std
```

```python
def cohenEffectSize(group1, group2):
    diff = group1.mean() - group2.mean()
    var1 = group1.var()
    var2 = group2.var()
    n1, n2 = len(group1), len(group2)
    pooled_var = (n1 * var1 + n2 * var2) / (n1 + n2)
    d = diff / math.sqrt(pooled_var)
    return d

d = cohenEffectSize(firsts.prglngth, others.prglngth)
d
```

```python
def Mode(hist):
    d = {k:hist[k] for k in hist}
    return max(d, key=d.get)

def AllModes(hist):
    d = {k:hist[k] for k in hist}
    alist = list()
    
    for k in d:
        alist.append((d[k], k))
    
    slist = sorted(alist, reverse=False)
    return slist
    
var = AllModes(firsts_hist)
var
    
```
